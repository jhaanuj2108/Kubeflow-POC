```python
from __future__ import print_function
import mlflow
#Set the tracking uri as Databricks but only after you configure databricks and dbfs
mlflow.set_tracking_uri("databricks")
mlflow.tracking.get_tracking_uri()
```

```python
#Write a experiment name that you want to load or write a new experiment name so that a new wxperiment will be created in databricks mlflow
experiment_name = "/Users/<your_username>/<your_experiment_name>"
run_origin = "jupyter-lab"
```

```python
#Here you will get various information like the mlflow version that you are running, experiment id for the experiment name that you provided earlier.
print("MLflow Version:",mlflow.version.VERSION)
mlflow.set_experiment(experiment_name)
mlflow_client = mlflow.tracking.MlflowClient()
experiment_id = mlflow_client.get_experiment_by_name(experiment_name).experiment_id
print("experiment_id:",experiment_id)
#print("experiment_name:",experiment_name)
```

```python
#A function to get the run_id of the last run experiment.
def get_last_run_info(experiment_id):
    infos = mlflow_client.list_run_infos(experiment_id)
    infos = sorted(infos, key=lambda x: x.start_time, reverse=True)
    info = infos[0]
    run_id = info.run_uuid
    return run_id
```

```python
run_id = get_last_run_info(experiment_id)
```
